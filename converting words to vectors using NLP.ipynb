```python
sw = stopwords.words("english")
```

* saving all the stopwords 

```python
b_n=[]
for i in data["review_text"]:
    l=[]
    ls=i.lower()
    ls=re.sub("[^a-zA-Z ]","",ls)
    ls=emoji.demojize(ls)
    for word in word_tokenize(ls):
        if word in sw:
            pass
        else:
            stems=PorterStemmer().stem(word)
            l.append(stems)
    b_n.append(" ".join(l))
```

* text pre-processing steps

```python
d = pd.DataFrame({"cleaned_data":b_n})
```

* after completion of pre-processing i have get cleaned data

```python
d1=d.head()
```

* this is the cleaned data

## BAG OF WORDS VECTORS (COUNT VECTORIZER)

```python
cv = CountVectorizer()
t_array = cv.fit_transform(d1.cleaned_data).toarray()
vocabulary = cv.vocabulary_
```

```python
pd.DataFrame(t_array,columns=sorted(vocabulary))
```
